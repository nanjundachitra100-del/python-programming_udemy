```python
class shape:
    def __init__(self):
        print(shape.area())

class Rectangle(shape):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    
    def area(self):
        return self.length * self.breadth
    
class Circle(shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.14 *self.radius*self.radius
    
rect=Rectangle(4,5)
circle=Circle(5)

print(rect.area())
print(circle.area())
```
