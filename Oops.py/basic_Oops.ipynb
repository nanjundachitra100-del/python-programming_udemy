```python
class dog:
    def __init__(self,color,breed):
        self.color=color
        self.breed=breed

    def dogy(self):
        print(f"Dog is {self.color} and is of {self.breed} breed")
       

dog1=dog("golden","lab")
dog2=dog("grey","pug")

dog1.dogy()
dog2.dogy()

```

```python
#modeling bank account
class Bank:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f"{self.balance}")

    def withdraw(self,amount):
        if self.balance<amount:
            print("Insufficient balance")
        else:
            self.balance_=amount
            print(f"{self.balance}")

    def get_balance(self):
        return self.balance
    
account=Bank("chitra",4000)

account.deposit(200)
account.withdraw(3000)
```
