```python
class dog:
    def __init__(self,color,breed):
        self.color=color
        self.breed=breed

    def dogy(self):
        print(f"Dog is {self.color} and is of {self.breed} breed")
       

dog1=dog("golden","lab")
dog2=dog("grey","pug")

dog1.dogy()
dog2.dogy()

```

```python
#modeling bank account
class Bank:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f"{self.balance}")

    def withdraw(self,amount):
        if self.balance<amount:
            print("Insufficient balance")
        else:
            self.balance_=amount
            print(f"{self.balance}")

    def get_balance(self):
        return self.balance
    
account=Bank("chitra",4000)

account.deposit(200)
account.withdraw(3000)
print(account.get_balance)
```

```python
#car models
class car:
    def __init__(self,brand,make,year):
        self.brand=brand
        self.make=make
        self.year=year
        
    def cars(self):
        print(f"{self.brand} by {self.make} in {self.year}")

    def start_engine(self):
         print(f"The {self.brand} engine has started!")




car1=car("kvu","toyato",1999)

car1.cars()
#calling a method
car1.start_engine()
        
```

```python
#inheritence
class bike(car):
    pass

bike1=bike("ktm","hero",2010)

bike1.cars()
bike1.start_engine()
```
