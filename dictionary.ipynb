```python
dict={
    "name":"chitra",
    "age":25,
    "city":"bangalore"
}
print(dict)
```

```python
#accesing the elements
print(dict["name"])
print(dict["age"])
```

```python
#modifying 
dict["age"]=20
print(dict["age"])
```

```python
print(dict.keys())
```

```python
#using for loop to iterate
for key in dict.keys():
    print(key)
print("\n")
for value in dict.values():
    print(value)
```

```python
students={
    "student1":{"name":"chitra","age":25},
    "student2":{"name":"karthik","age":20}
}
print(students)
print(students["student1"]["name"])
print(students["student2"]["name"])

for key ,value in students.items():
    print(key,value)
```
