```python
#example 1
def even(num):
    if num %2==0:
        return True
lst=[1,3,2,4,5,6,7,8,9]
list(filter(even,lst))
```

```python
#example 2
def num(n):
    if n<5:
        return True
lst=[1,2,3,4,5,6,7,8,90]
list(filter(num,lst))
```

```python
#filter with lambda function
lst=[1,2,3,4,5]
odd=list(filter(lambda x:x%2!=0,lst))
print(odd)
```
