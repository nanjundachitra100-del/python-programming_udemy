```python
#function with no arguments and no return type
def greet():
    print("Hello, World!")
greet()
```

```python
#functins with argumnet and no return type
def add(a,b):
    print("sum is:",a+b)
add(5,3)
```

```python
#functions with arguments and return type
def add(a,b):
    sum_result=a+b
    return sum_result
result=add(5,6)
print(result)
```

```python
#variable length argument
def add(*args):
    print(args)
add(1,2,3,4)

def kwargs(**kwargs):
    print(kwargs)
kwargs(name="chitra",age=20,city="bangalore")
    
```
