```python
#function copy
def welcome():
    print("welcome darling")

welcome()

wel=welcome
wel()
del welcome
```

```python
#closures- closures is a inner function that stores the variables of outer function
def main_welcome():
    msg="welcome"
    
    def sub_welcome():
        print("welcome to my home")
        print("feel comfortable")
        print(msg)
    return sub_welcome()
main_welcome()
```

```python
def main_welcome(msg):
    
    def sub_welcome():
        print("welcome to my home")
        print("feel comfortable")
        print(msg)
    return sub_welcome()
main_welcome("welcome")
```

```python
#decorators
def main_welcome(msg):
    
    def sub_welcome():
        print("welcome to my home")
        print("feel comfortable")
        msg()
    return sub_welcome


@main_welcome
def greet():
    print("everyone be carefull for ur belongings")
    
```
