```python
#without generators (using return)
def num():
    return [1,2,3,4,5]

x=num()
print(x)
```

```python
#using generators(yield)
def num():
    yield 1
    yield 2
    yield 3
    yield 4
    yield 5

x=num()
print(next(x))
print(next(x))
print(next(x))

#this saves the memoery and gives output one after the other only if needed unlike return
    
```
