```python
num=[1,2,3,4,5]
x=iter(num) #creates an iterator

print(next(x)) #moves to next iterator
print(next(x))
print(next(x))
print(next(x))
print(next(x))


# exceptions 
try:
    print(next(x))
except StopIteration:
    print("there is no element to iterate")
```

```python
#using for loop
num=[1,2,3,4,5]
x=iter(num)

for i in x:
    print(i)
```

```python
#using while loop
num=[1,2,3,4,5]
x=iter(num)

while True:
    try:
        print(next(x))
    except StopIteration:
        print("there is no element to iterate")
        break
```
