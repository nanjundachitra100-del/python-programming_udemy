```python
lst=[]
for i in range(10):
    lst.append(i)
    print(lst)
```

```python
lst=[]
lst = [x**2 for x in range(10)]
print(lst)
```

```python
lst=[x for x in range(10) if x%2==0]
print(lst)
```

```python
#nested list comprehension
lst1=[1,2,3,4]
lst2=['a','b','c']
pair=[[i,j]for i in lst1 for j in lst2]
print(pair)
```

```python
#with if else statment
lst=[x if x%2==0  else 'odd' for x in range(10)]
print(lst)
```

```python
#nested list 3x3 matrix
lst1=[1,2,3]
lst2=[4,5,6]
lst3=['a','b','c']
matrix=[[i,j,k] for i in lst1 for j in lst2 for k in lst3]
print(matrix)
```

```python
#transpose of matrix
matrix=[[1,2,3],[4,5,6],[7,8,9]]
transpose=[]
for i in range(len(matrix[0])):
    new_row=[]
    for row in matrix:
        new_row.append(row[i])
        transpose.append(new_row)
print(transpose)
```

```python
matrix=[[1,2,3],[4,5,6],[7,8,9]]
transpose=[[row[i]for row in matrix]for i in range(len(matrix[0]))]
print(transpose)
```

```python
#list intersection
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

intersection = []

for item in list1:
    if item in list2:
        intersection.append(item)

print(intersection)
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

intersection=list(zip(list1,list2))
print(intersection)
```
