```python
#creating a set
my_set={1,2,3,4,5}
print(my_set)
```

```python
my_set.add(6)
print(my_set)
```

```python
my_set.update([1,3,7])
print(my_set)
```

```python
my_set.remove(6)
my_set.remove(7)
print(my_set)
```

```python
my_set.pop()
print(my_set)
```

```python
my_set.clear()
print(my_set)
```

```python
set1={1,2,3,4,5}
set2={3,4,5,6,7}
print(set1.union(set2))
```

```python
#combines both the sets and removes the duplicate elements
print(set1.intersection(set2))
```

```python
#membership operation
print(3 in set1)
print(11 in set1)
```

```python
#gives comman elements and update it to set1 
set1={1,2,3,4,5}
set2={3,4,5,6,7}
print(set1.intersection_update(set2))
print(set1)
```
