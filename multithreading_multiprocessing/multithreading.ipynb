```python
#multithreading
import threading
import time

#creating 2 functions or 2 tasks
def print_numbers():
    for i in range(5):
        time.sleep(2)
        print(f"Numbers:{i}")

def print_letters():
    for letter in "ABCDEF":
        time.sleep(2)
        print(f"Letters:{letter}")

#creating threads
t1=threading.Thread(target=print_numbers)
t2=threading.Thread(target=print_letters)

t=time.time()

#start the thread
t1.start()
t2.start()

t1.join()
t2.join()

finished_time=time.time()-t
print(finished_time)
```
